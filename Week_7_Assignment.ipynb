# Week 7: Inferential Statistics â€“ Hypothesis Testing

This notebook explores key concepts in inferential statistics, focusing on hypothesis testing.

## Population vs Sample

*   **Population:** The entire group of individuals or objects that you are interested in studying. It's the complete set of data points.
*   **Sample:** A subset of the population that is selected for study. It's often impractical or impossible to study the entire population, so we use samples to draw conclusions about the population.

## Sampling Techniques

Various methods are used to select a sample from a population. The goal is to obtain a sample that is representative of the population to avoid bias. Some common techniques include:

*   **Simple Random Sampling:** Every member of the population has an equal chance of being selected.
*   **Stratified Sampling:** The population is divided into subgroups (strata), and then random samples are taken from each stratum.
*   **Cluster Sampling:** The population is divided into clusters, and then a random sample of clusters is selected, with all members of the selected clusters included in the sample.
*   **Systematic Sampling:** Every nth member of the population is selected after a random starting point.

## Hypothesis Testing: Null & Alternate Hypotheses

Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample to draw conclusions about a population. It involves setting up two competing hypotheses:

*   **Null Hypothesis (Hâ‚€):** A statement of no effect or no difference. It's the hypothesis that the researcher is trying to disprove.
*   **Alternate Hypothesis (Hâ‚ or Ha):** A statement that contradicts the null hypothesis. It's the hypothesis that the researcher is trying to prove.

The goal of hypothesis testing is to determine whether to reject the null hypothesis in favor of the alternate hypothesis, or fail to reject the null hypothesis.

## p-values, Significance Levels (Î±)

*   **p-value:** The probability of observing a test statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming the null hypothesis is true. A small p-value suggests that the observed data is unlikely under the null hypothesis.
*   **Significance Level (Î±):** A threshold probability that is set before conducting the hypothesis test. It represents the maximum probability of rejecting the null hypothesis when it is actually true (Type I error). Common significance levels are 0.05 (5%) and 0.01 (1%).

If the p-value is less than or equal to the significance level (p â‰¤ Î±), we reject the null hypothesis. If the p-value is greater than the significance level (p > Î±), we fail to reject the null hypothesis.

## T-tests & Chi-Square Test (Intuition only, no heavy math)

These are statistical tests used in hypothesis testing.

*   **T-tests:** Used to compare the means of two groups. For example, you could use a t-test to see if there is a significant difference in the average test scores of two different teaching methods. The intuition is to see how far apart the means are relative to the variability within the groups.
*   **Chi-Square Test:** Used to examine the relationship between two categorical variables. For example, you could use a chi-square test to see if there is a significant association between gender and preference for a certain product. The intuition is to compare the observed frequencies in each category to the frequencies that would be expected if there were no association between the variables.

### Q1. A coin is tossed 50 times and shows 30 heads. State the null and alternate hypotheses to check if the coin is fair.


```python
# Week 7: Hypothesis Testing - Q1
# A coin is tossed 50 times and shows 30 heads.
# Test if the coin is fair using a one-sample proportion z-test.

import statsmodels.api as sm

# Given data
n = 50        # total tosses
x = 30        # number of heads
p0 = 0.5      # hypothesized probability for a fair coin

# Calculate the sample proportion
p_hat = x / n

# Perform the one-sample proportion z-test
z_stat, p_value = sm.stats.proportions_ztest(count=x, nobs=n, value=p0)

print(f"Sample proportion (pÌ‚): {p_hat:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("âŒ Reject the Null Hypothesis: The coin is not fair.")
else:
    print("âœ… Fail to Reject the Null Hypothesis: The coin can be considered fair.")
```

### Q2. A factory claims its bulbs last 1000 hours on average. Formulate the null and alternate hypotheses for testing this claim.

```python
# Q2. A factory claims its bulbs last 1000 hours on average.
# Formulate and test the hypothesis using sample data.

import numpy as np
from scipy import stats

# Sample data (for example)
sample = np.array([980, 1005, 995, 1020, 1010, 990, 1008, 1002, 997, 1015])

# Population mean (claimed)
mu_0 = 1000

# Sample mean and standard deviation
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
n = len(sample)

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(sample, mu_0)

print(f"Sample Mean = {sample_mean}")
print(f"T-Statistic = {t_stat}")
print(f"P-Value = {p_value}")

# Significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis (The claim may not be true).")
else:
    print("Fail to Reject the Null Hypothesis (The claim stands).")
```

### Q3. A new teaching method is applied to a class. Formulate hypotheses to test whether the new method improves average student marks.

```python
# Q3. A new teaching method is applied to a class.
# Test whether the new method improves average student marks.

import numpy as np
from scipy import stats

# Sample data: marks of students taught using the new method
new_method_marks = np.array([78, 82, 85, 80, 88, 90, 84, 86, 83, 89])

# Population mean (previous average marks before new method)
mu_0 = 80

# Perform one-sample t-test (right-tailed test since we want to check for improvement)
t_stat, p_value = stats.ttest_1samp(new_method_marks, mu_0)

print(f"Sample Mean = {np.mean(new_method_marks)}")
print(f"T-Statistic = {t_stat}")
print(f"P-Value = {p_value / 2}")  # Divide by 2 for one-tailed test

alpha = 0.05

if (p_value / 2) < alpha and t_stat > 0:
    print("Reject Null Hypothesis: The new method improves average marks.")
else:
    print("Fail to Reject Null Hypothesis: No significant improvement observed.")
```

### Q4. A company claims 40% of customers prefer online shopping. Formulate hypotheses to test this claim.



```python
# Import libraries
from statsmodels.stats.proportion import proportions_ztest

# Sample data
n = 200          # Total number of customers surveyed
x = 90           # Number of customers who prefer online shopping
p0 = 0.40        # Claimed proportion

# Perform Z-test
stat, p_value = proportions_ztest(count=x, nobs=n, value=p0)

print("Z-statistic:", stat)
print("P-value:", p_value)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Claim may not be true.")
else:
    print("Fail to reject the null hypothesis: Claim is consistent with data.")
```

### Q9. You have the heights of 10 male and 10 female students. Which test will you perform to check if their average heights are different?


```python
from scipy import stats

# Example data: heights in cm
male_heights = [170, 172, 168, 175, 169, 174, 171, 173, 176, 170]
female_heights = [160, 162, 158, 165, 159, 164, 161, 163, 166, 160]

# Perform independent two-sample t-test
t_stat, p_value = stats.ttest_ind(male_heights, female_heights)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Average heights are significantly different.")
else:
    print("Fail to reject H0: No significant difference in average heights.")
```

### Q10. A training program is introduced. Employee performance is measured before and after. Which T-test will check if the program improved performance?


```python
from scipy import stats

# Example data: performance scores
before_training = [70, 68, 75, 65, 72, 69, 74, 71, 73, 70]
after_training = [75, 70, 78, 70, 76, 73, 77, 74, 76, 75]

# Perform paired t-test
t_stat, p_value = stats.ttest_rel(after_training, before_training)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value/2 < alpha and t_stat > 0:  # one-tailed test for improvement
    print("Reject H0: The training program significantly improved performance.")
else:
    print("Fail to reject H0: No significant improvement from training.")
```

### Q11. You roll a dice 120 times and record how many times each number appears. Use a Chi-Square test to check if the dice is fair.

```python
from scipy.stats import chisquare

# Observed counts from rolling dice 120 times
observed_counts = [18, 22, 20, 19, 21, 20]  # Example data

# Expected counts if the dice is fair
expected_counts = [120/6] * 6

# Perform Chi-Square Goodness-of-Fit Test
chi_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

print("Chi-Square Statistic:", chi_stat)
print("P-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The dice is not fair.")
else:
    print("Fail to reject H0: The dice is fair.")
```

### Q12. A chips brand claims the average packet weight is 50g. You weight 10 random packets:
` [49, 52, 51, 50, 48, 49, 50, 51, 52, 50] `
 ### Use a One-Sample T-Test to check if the claim is true.

```python
from scipy import stats

# Sample data: weights in grams
weights = [49, 52, 51, 50, 48, 49, 50, 51, 52, 50]

# Claimed mean
claimed_mean = 50

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(weights, claimed_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The claim is not true. Mean weight differs from 50g.")
else:
    print("Fail to reject H0: The claim is consistent with the data.")
```

### Q13. A hospital claims a new medicine has a 70% success rate. In a trial of 20 patients, 12 recovered. Use a proportion test to check if the success rate matches the claim.

```python
from statsmodels.stats.proportion import proportions_ztest

# Sample data
n = 20        # number of patients
x = 12        # number of successes
p0 = 0.7      # claimed proportion

# Perform one-sample proportion Z-test
stat, p_value = proportions_ztest(count=x, nobs=n, value=p0)

print("Z-statistic:", stat)
print("P-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The success rate differs from the claimed 70%.")
else:
    print("Fail to reject H0: The success rate is consistent with the claim.")
```

### Q14. A survey of 30 people records gender (Male/Female) and favorite drink (Tea/Coffee). Perform a Chi-Square test to check if drink preference depends on gender.


```python
from scipy.stats import chi2_contingency
import numpy as np

# Example survey data
# Rows: Gender (Male, Female)
# Columns: Drink (Tea, Coffee)
# Suppose the counts are:
# Male: 8 Tea, 7 Coffee
# Female: 6 Tea, 9 Coffee

data = np.array([[8, 7],
                 [6, 9]])

# Perform Chi-Square Test of Independence
chi2_stat, p_value, dof, expected = chi2_contingency(data)

print("Chi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)
print("Expected Frequencies:\n", expected)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: Drink preference depends on gender.")
else:
    print("Fail to reject H0: Drink preference is independent of gender.")
```

### Q15. A factory claims its bulbs last 1000 hours on average. You test 8 bulbs:
`[980, 1005, 995, 1010, 1000, 985, 1020, 990]`
### Use a One-Sample T-Test to verify the claim.

```python
from scipy import stats

# Sample data: lifespans in hours
lifespans = [980, 1005, 995, 1010, 1000, 985, 1020, 990]

# Claimed mean
claimed_mean = 1000

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(lifespans, claimed_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The claim is not true. Mean lifespan differs from 1000 hours.")
else:
    print("Fail to reject H0: The claim is consistent with the data.")
```
